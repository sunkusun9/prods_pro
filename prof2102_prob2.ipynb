**Step B-1.**

prof2102_2_personnel_train.csv (총 13칼럼)에는 범주형 데이터를 가지는 

Department, Education, Gender, JobRole, MaritalStatus 칼럼들(5개)이 있습니다. 

나머지 8개의 수치형 칼럼을 포함하고, 5개의 범주형 칼럼은 One-hot encoding으로 변환한 데이터셋 EMP_P1(총 30칼럼)을 만드시오. 

One-hot encoding 시, 변환되는 칼럼의 수는 범주의 수와 같도록 하시오.

````python
df_emp_p1 = pd.get_dummies(
    df_pers_train, 
    columns=['Department', 'Education', 'Gender', 'JobRole', 'MaritalStatus'], # 가변수 대상 컬럼을 전달합니다.
    drop_first=False # One-hot encoding 시, 변환되는 칼럼의 수는 범주의 수와 같도록 합니다.
)
````

**Step B-2.**

prof2102_2_work_record.csv 데이터셋으로부터 각 직원 ID(EmpID) 별로 일 평균 근무 시간 정보를 가지는 

EMP_P2 데이터셋을 다음 설명을 참고하여 생성합니다. 

이 데이터셋은 근로자ID(변수명: EmpID), 4월 1일부터 7월 31일까지의 일 평균 근무 시간(변수명: WorkMean) 2개 컬럼을 갖는다.

<그림 1>에서 나타낸 바와 같이, prof2102_2_work_record.csv 데이터셋에서 근로자 별로 4월 1일부터 7월 31일까지 

근무시간의 평균을 계산하여 EMP_P2 데이터셋의 WorkMean 칼럼 값을 계산합니다. 휴가 등으로 인해 근로하지 않은 경우가 있어 

prof2102_2_work_record.csv 데이터셋에 결측값(NaN/null)이 존재할 수 있습니다.

결측값 모두 0으로 대체 후 평균 근로 시간을 계산 하시오(EMP_P2의 총 행 수는 4,400개입니다). 

평균 계산 시 값이 0인 날짜를 포함하여 모든 날짜를 사용합니다.

image.png
<그림 1> EMP_P2 데이터셋의 생성

````python
# 결측값 모두 0으로 대체합니다. ※ 결측값은 평균계산시 카운트 되지 않습니다.
# WorkMean으로 이름을 대체합니다.
s_emp_p2 = df_work.fillna(0).mean(axis=1).rename('WorkMean')
s_emp_p2
````

**Step B-3.**

EMP_P1과 EMP_P2 데이터셋을 조인합니다. 즉, EmpID가 같은 행들을 결합합니다. 

조인 결과 데이터셋을 최종 EMP_TRAIN 데이터셋이라 하겠습니다. 

EMP_TRAIN의 행들은 EmpID의 오름차순으로 정렬되어 있어야 합니다.

````python
# Join은 호출자에서는 Index와 Column을 모두 기준으로 할 수 있고, 피호출자는 Index를 기준으로 합니다.
# 결합의 기준 컬럼이 모두 Index에 위치해 있으므로, join 만으로 결합이 성립합니다.
df_emp_train = df_emp_p1.join(s_emp_p2).sort_index()
df_emp_train
````

````python
# 상황 1
# 다른 상황을 만들어 봅니다
# 호출자에서 Index에 위치한 EmpID를 컬럼에 위치 시킵니다.
# 이때는 on을 지정해야만 합니다.
df_emp_p1.reset_index().join(s_emp_p2, on='EmpID')
````

````python
# 상황 2: 결합기준이  피호출자 일반 컬럼에 있을 경우
# merge는 피호출자가 Index(인덱스 이름이 있어합니다.), Column 있을 경우 모두 지원합니다.
df_emp_p1.merge(s_emp_p2.reset_index(), on='EmpID')
````

## 문제 5번  10점

**(EMP_TRAIN, EMP_TEST, SALARY_TRAIN, SALARY_TEST를 이용하여)** 
이번 문제에서는 직원들의 정보를 바탕으로 합리적인 급여를 산출할 수 있는 모델을 만들고자 합니다.

EMP_TRAIN과 EMP_TEST 데이터셋의 인적 정보를 활용합니다.

EMP_TRAIN과 EMP_TEST의 각 행에 해당하는 근로자의 월 급여는 SALARY_TRAIN과 SALARY_TEST의 MonthlySalary 칼럼에 있습니다. 

EMP_TRAIN 데이터셋과 SALARY_TRAIN의 MonthlySalary 칼럼을 사용하여 근로자의 월 급여를 예측하는 Regression 모델을 학습합니다.

EmpID 칼럼을 제외한 EMP_TRAIN의 모든 칼럼(총 30개)을 feature로, SALARY_TRAIN의 '현재 월 급여(MonthlySalary)' 칼럼을 label로 하는 Regression모델을 학습합니다. 

(즉, 3,500개의 데이터를 이용하여 학습)

모델은 여러 Decision Tree들을 결합한 Ensemble 모델인 Random Forest를 사용합니다. 모델 학습 시 다음의 파라미터 조건을 따르시오.

**Python 함수가이드**
```
from sklearn.ensemble import RandomForestRegressor
- n_estimators = 10
- random_state = 42
문제 지시 사항 외 Default 파라미터 사용
```

테스트 데이터인 EMP_TEST와 SALARY_TEST의 900개 행 중, 처음 500개의 행만 활용하여 모델을 평가합니다. 

(나머지 400개의 행의 경우 MonthlySalary가 0입니다. 추후 문제 8에서 400개의 행에 대한 MonthlySalary 값을 예측하고 제출하여야 합니다.) 

모델 평가는 Mean absolute percentage error (MAPE)를 사용합니다. MAPE는 다음과 같이 정의됩니다.

$MAPE=\frac{1}{n}\sum_{i=1}^{n}|\frac{(actual_i-predicted_i)}{actual_i}| × 100 $ (n=500)

$actual_i$: i 번째 행의 실제 “MonthlySalary” 값

$predicted_i$: i 번째 행의 예측된 “MonthlySalary” 값

학습된 모델을 테스트 데이터(즉, 처음 500개 행)에 적용하여 Mean absolute percentage error (MAPE)를 계산하고, 

소수점 셋째 자리에서 반올림하여 둘째 자리까지 기술하시오.

 (답안예시) 12.34


````python
# EMP_TRAIN, EMP_TEST와 순서를 맞추기 위해 EmpID(Index)에 대해 정렬이 되도록합니다.
df_salary_train.sort_index(inplace=True)
df_salary_test.sort_index(inplace=True)
````

````python
from sklearn.ensemble import RandomForestRegressor
# RandomForestRegressor 주요 Hyper-Parameter 
# n_estimators: 트리의 수 (앙상블 모델의 수)
# max_depth: 트리의 최대 깊이
# min_samples_split: 노드 분기에 필요한 최소 샘플 수
# min_samples_leaf: 단말 노드의 최소 샘플 수
# bootstrap: True: 복원추출사용

# MAPE는 직접 구현해야 합니다.
def mape(y, y_hat):
    return np.mean(np.abs(y - y_hat) / y) * 100

reg_rf = RandomForestRegressor(
    n_estimators = 10,
    random_state = 42
)
reg_rf.fit(df_emp_train, df_salary_train['MonthlySalary'])
````

````python
(
    mape(df_salary_train['MonthlySalary'], reg_rf.predict(df_emp_train)), # Train의 성능
    mape(df_salary_test.iloc[:500]['MonthlySalary'], reg_rf.predict(df_emp_test.iloc[:500])) # Test의 성능
)
````

Principle Component Analysis (PCA)를 활용하여 outlier를 찾고, 이를 예측에 활용하고자 합니다. 다음 단계를 순서대로 수행하시오.

````python
from sklearn.decomposition import PCA
pca = PCA(
   n_components=0.9 # 기존 변수의 variance 정보를 90% 이상 설명할 수 있는 최소 차원의 수 N
)
X_pca = ['JobInvolvement', 'PerformanceRating', 'DepartmentFitness', 'WorkLifeBalance', 'PercentSalaryHike']
pca.fit(df_eval_train[X_pca])
````

**Step 6-2.**

'N 차원으로 축소된 PCA 결과(PCA results)' 즉, N개의 컬럼으로 표현된 projected values를 기존 5차원 데이터의 형태로 

아래와 같이 재구축(reconstruction)합니다. 

$PCA reconstruction = (PCA results) × (Eigenvectors)^⊤  + (Data 평균)$

$Table size (nx5) = (nxN) × (Nx5) + (nx5)$, n=number of rows in X


PCA().components_ 사용하여 $(Eigenvectors)^⊤$  계산

````python
# 방법 #1
X_hat = np.dot(pca.transform(df_eval_train[X_pca]), pca.components_) + pca.mean_
X_hat
````

**Step 6-3.** 

단계 2에서 재구축된 데이터셋을 $\hat{X}$이라고 하겠습니다. X와 $\hat{X}$의 거리를 기반으로 Outlier를 정합니다. 

$X$와 $\hat{X}$의 i 번째 데이터 사이의 거리는 $d_i≔(X_i-\hat{X}_i )^⊤ (X_i- \hat{X}_i )$ 을 사용합니다. (즉, $d_i$는 유클리드 거리의 제곱입니다.) 

이 값은 각 행당 1개씩 생성되며, 이 값을 기준으로 outlier를 판단합니다. 

특정 threshold τ에 대해, $d_i>\tau$ 이면 데이터를 outlier로 판단하여 

EVALUATION_TRAIN, SATISFACTION_TRAIN 데이터셋에서 제외합니다. 

(Threshold $\tau$는 1부터 4 까지의 정수들 중 하나로 선정할 예정
